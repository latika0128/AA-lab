```python
def topologicalSortUtil(v_node, adj, visited, stack):
    visited[v_node] = True
    for i in adj[v_node]:
        if not visited[i]:
            topologicalSortUtil(i, adj, visited, stack)
    stack.append(v_node)

def topologicalSort(v_count, edges):
    # Build adjacency list
    adj = [[] for _ in range(v_count)]
    for src, dest in edges:
        adj[src].append(dest)

    visited = [False] * v_count
    stack = []

    # Call the recursive helper function for all vertices one by one
    for i in range(v_count):
        if not visited[i]:
            topologicalSortUtil(i, adj, visited, stack)

    # The stack contains the topological order in reverse.
    # Pop elements to get the correct topological order.
    result = []
    while stack:
        result.append(stack.pop())
    print("Topological Sort (DFS):", result)

# Example Usage:
v_count = 6
edges = [(5,0), (4,0), (4,1), (3,1), (2,3), (5,2)]
topologicalSort(v_count, edges)
```

```python
from collections import deque

v = 6
edges = [[5,0], [4,0], [4,1], [3,1], [2,3], [5,2]]

def kahns_topological_sort(v, edges):
    adj = [[] for _ in range(v)]
    in_degree = [0] * v

    # Build adjacency list and in-degree array
    for u, w in edges:
        adj[u].append(w)
        in_degree[w] += 1

    print("Initial in-degree:", in_degree)

    # Queue for all vertices with in-degree 0
    queue = deque([i for i in range(v) if in_degree[i] == 0])

    top_order = []

    # Process vertices
    while queue:
        u = queue.popleft()
        top_order.append(u)

        for neighbor in adj[u]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    # Check if topological sort is possible (no cycle)
    if len(top_order) == v:
        return top_order
    else:
        return "Graph has a cycle, topological sort not possible."

# Run and print the topological sort order
result = kahns_topological_sort(v, edges)
print("Topological sort order:", result)
```
