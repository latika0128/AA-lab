```python
import sys

def tsp(cost, n):
    # Initialize DP table with infinity
    dp = [[sys.maxsize] * n for _ in range(1 << n)]
    dp[1][0] = 0  # Starting from city 0 with only city 0 visited

    # Fill DP table
    for mask in range(1 << n):  # Iterate over all subsets of cities
        for i in range(n):  # Iterate over each city
            if mask & (1 << i):  # City i is visited in subset mask
                for j in range(n):  # Check all other cities in the subset
                    if mask & (1 << j) and i != j:  # Exclude city i
                        dp[mask][i] = min(dp[mask][i], dp[mask ^ (1 << i)][j] + cost[j][i])

    # Find minimum cost path returning to start (city 0)
    result = sys.maxsize
    for i in range(1, n):
        result = min(result, dp[(1 << n) - 1][i] + cost[i][0])

    return result

# Example usage:
if __name__ == "__main__":
    # Example cost matrix (symmetric)
    # cost[i][j] represents the cost of traveling from city i to city j
    cost = [
        [0, 10, 15, 20],  # City 0
        [10, 0, 35, 25],  # City 1
        [15, 35, 0, 30],  # City 2
        [20, 25, 30, 0]   # City 3
    ]
    n = len(cost)

    # Solve TSP
    result = tsp(cost, n)
    print(f"The minimum cost to complete the tour is: {result}")
```
